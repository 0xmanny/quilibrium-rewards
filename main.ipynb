```python
import pandas as pd
import json
```

```python
# read json
data = json.load(open('data.json','rb'))

# convert to data frame
df = pd.DataFrame(data)

# fill NaN with 0
df.fillna(0, inplace=True)

# group to aggregate rewards (in case duplicates)
s = df.groupby('peerId')['reward'].sum()

# filter out zero rewards
s = s[s > 0]
```

```python
# plot histogram
s.plot(kind='hist', bins=100)
```

```python
# descriptive statistics 
s.describe()
```
